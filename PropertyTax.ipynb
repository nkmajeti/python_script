```python
import urllib.request
from bs4 import BeautifulSoup

def get_link_list():
    state_list_url = "https://smartasset.com/taxes/property-taxes"
    state_list_page = urllib.request.urlopen(state_list_url)
    soup = BeautifulSoup(state_list_page, features="html.parser")

    links = []
    for ul in soup.findAll('ul', class_='related_link_list'):
        for link in ul.findAll('a'):
            if link.has_attr('href') and link['href'].startswith('/taxes'):
                links.append(link['href'])
    return links

def read_tax_data():
    links = get_link_list()
    for link in links:
        print(link)
        tax_page = urllib.request.urlopen("https://smartasset.com" + link)

        soup = BeautifulSoup(tax_page, features="html.parser")
        tax_div = soup.find("div", {"id": "table1-tab1"})
        if tax_div is not None:
            tax_table = tax_div.find('table')
            link_list = []

            for row in tax_table.findAll("tr"):
                cells = row.findAll('td')
                row = [row.text for row in cells]
                link_list.append(row)
                print(row)

            
            
read_tax_data()
```
